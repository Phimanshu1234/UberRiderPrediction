```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(input):
    return 1.0/(1+np.exp(-input))
```

```python
z=np.linspace(-8,8,1000)
y=sigmoid(z)
plt.plot(z,y)
plt.yticks([0.0,0.25,0.50,0.75,1.0])
plt.axhline(y=0,ls='dotted',color='k')
plt.axhline(y=0.50,ls='dotted',color='k')
plt.axhline(y=1.0,ls='dotted',color='k')
plt.title('Logistic Regression Core Implementation')
plt.xlabel('z')
plt.ylabel('y(z)')
plt.show()
```
